```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
df.head()
```

Droping column salary_more_then_100k 

```python
inputs = df.drop('salary_more_then_100k', axis = 'columns')
target = df['salary_more_then_100k']
```

```python
target
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

Encoders Loading

```python
company = LabelEncoder()
job = LabelEncoder()
degree = LabelEncoder()
```

Encoding columns
https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html

```python
inputs['company_n'] = company.fit_transform(inputs['company'])
inputs['job_n'] = company.fit_transform(inputs['job'])
inputs['degree'] = company.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['company','job'],axis = 'columns')
```

```python
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

Fitting the dataset using decisionTree algo

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[1,2,2]])
```
